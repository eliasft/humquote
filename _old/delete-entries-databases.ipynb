```python
# In[]:
import sqlite3
import pandas as pd

# Database path
db_path = r"C:\Users\ftelias\OneDrive\Documents\GitHub\humquote\futures_prices.db"

# Connect to the database
conn = sqlite3.connect(db_path)

# First, let's look at the current data
print("Current data in the database:")
df = pd.read_sql_query("SELECT * FROM futures_data", conn)
print(df)

# In[]:
# Delete entries for the specified dates
dates_to_delete = ['2024-11-18','2024-11-18']
delete_query = """
DELETE FROM futures_data 
WHERE `Quote Date` IN (?, ?)
"""

try:
    cursor = conn.cursor()
    cursor.execute(delete_query, dates_to_delete)
    conn.commit()
    print(f"\nDeleted {cursor.rowcount} rows.")
except sqlite3.Error as e:
    print(f"An error occurred: {e}")

# In[]:
# Verify the deletion
print("\nVerifying remaining data:")
df_after = pd.read_sql_query("SELECT * FROM futures_data", conn)
print(df_after)

# In[]:
# Close the connection
conn.close()

# Optional: Display the unique dates remaining in the database
print("\nUnique dates remaining in the database:")
print(sorted(df_after['Quote Date'].unique()))
```

```python
# In[]:
import sqlite3
import pandas as pd

# Database paths
futures_db_path = r"C:\Users\ftelias\OneDrive\Documents\GitHub\humquote\futures_prices.db"
bulk_price_db_path = r"C:\Users\ftelias\OneDrive\Documents\GitHub\humquote\bulk_price_tracker.db"

# Dates to delete
dates_to_delete = ['2024-11-16', '2024-11-17',18']

# In[]:
# First handle futures_prices.db
print("Processing futures_prices.db...")
conn_futures = sqlite3.connect(futures_db_path)

# Show current data
print("\nCurrent data in futures_data table:")
df_futures = pd.read_sql_query("SELECT * FROM futures_data", conn_futures)
print(df_futures)

# Delete entries
delete_query_futures = """
DELETE FROM futures_data 
WHERE `Quote Date` IN (?, ?)
"""

try:
    cursor = conn_futures.cursor()
    cursor.execute(delete_query_futures, dates_to_delete)
    conn_futures.commit()
    print(f"\nDeleted {cursor.rowcount} rows from futures_data table.")
except sqlite3.Error as e:
    print(f"An error occurred with futures_prices.db: {e}")

# Verify deletion
print("\nVerifying remaining data in futures_data:")
df_futures_after = pd.read_sql_query("SELECT * FROM futures_data", conn_futures)
print(df_futures_after)

# Close futures database connection
conn_futures.close()

# In[]:
# Now handle bulk_price_tracker.db
print("\nProcessing bulk_price_tracker.db...")
conn_bulk = sqlite3.connect(bulk_price_db_path)

# Show current data
print("\nCurrent data in bulk_price_index table:")
df_bulk = pd.read_sql_query("SELECT * FROM bulk_price_index", conn_bulk)
print(df_bulk)

# Delete entries
delete_query_bulk = """
DELETE FROM bulk_price_index 
WHERE `Quote Date` IN (?, ?)
"""

try:
    cursor = conn_bulk.cursor()
    cursor.execute(delete_query_bulk, dates_to_delete)
    conn_bulk.commit()
    print(f"\nDeleted {cursor.rowcount} rows from bulk_price_index table.")
except sqlite3.Error as e:
    print(f"An error occurred with bulk_price_tracker.db: {e}")

# Verify deletion
print("\nVerifying remaining data in bulk_price_index:")
df_bulk_after = pd.read_sql_query("SELECT * FROM bulk_price_index", conn_bulk)
print(df_bulk_after)

# Close bulk price database connection
conn_bulk.close()

# In[]:
# Summary of unique dates remaining in both databases
print("\nUnique dates remaining in futures_data:")
print(sorted(df_futures_after['Quote Date'].unique()))

print("\nUnique dates remaining in bulk_price_index:")
print(sorted(df_bulk_after['Quote Date'].unique()))
```
